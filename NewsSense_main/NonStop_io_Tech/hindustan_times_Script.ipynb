```python
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
import time
```

```python
import pandas as pd
```

```python
driver=webdriver.Chrome()
driver.maximize_window()
driver.get('https://www.hindustantimes.com/')
```

```python
home_cric=driver.find_element(by='xpath',value="//ul[@class='leftFixedNav']//li/a[@href='https://www.hindustantimes.com/cricket']")
home_cric.click()
```

```python
data_container=driver.find_element(by='xpath',value="//section[@id='dataHolder']")
data=data_container.find_elements(by='xpath',value=".//div[@class='cartHolder listView track timeAgo']")
type(data),len(data)
```

```python
text_data_cric=[]
for i in data:
    text_data_cric.append(i.find_element(by='xpath',value=".//h3[@class='hdg3']/a").text)
```

```python
text_data_cric
```

```python
cric_data=pd.DataFrame({'lable':'cricket','text':text_data_cric})
```
