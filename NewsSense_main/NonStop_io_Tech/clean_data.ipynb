```python
import re
```

```python
def clean(text):
  clean_text=re.sub('[^a-zA-Z0-9]+',' ',text)
  text_lower=clean_text.lower()
  return text_lower
```

```python
clean("Disha Salian case: Indications of SIT inquiry out of BJP’s fear, says Aaditya Thackeray")
```

```python
data['clean_text']=data.text.apply(clean)
```

```python
text='disha salian case indications of sit inquiry out of bjp’s fear says aaditya thackeray'
```

```python
doc=nlp(text)

for token in doc:
  print(token.text , "==>",token.pos_)
```

```python
def prop_(text):
  doc=nlp(text)
  count=[i for i in doc if i.pos_=='PROPN']
  return len(count)
```

```python
def noun_(text):
  doc=nlp(text)
  count=[i for i in doc if i.pos_=='NOUN']
  return len(count)
```

```python
def adj_(text):
  doc=nlp(text)
  count=[i for i in doc if i.pos_=='ADJ']
  return len(count)
```

```python
def verb_(text):
  doc=nlp(text)
  count=[i for i in doc if i.pos_=='VERB']
  return len(count)
```

```python
data['prop_count']=data.clean_text.apply(prop_)
```

```python
data['noun_count']=data.clean_text.apply(noun_)
```

```python
data['adj_count']=data.clean_text.apply(adj_)
```

```python
data['verb_count']=data.clean_text.apply(verb_)
```

```python
def stop_(text):
  count=[i for i in nlp(text) if i.is_stop==True]
  return len(count)
```

```python
stop_(data.clean_text[1])
```

```python
data['stop_count']=data.clean_text.apply(stop_)
```

```python
data.head()
```

```python
def clean_token(text):
  doc=nlp(text)
  token_=[]
  for token in doc:
    token_.append(token.lemma_)
  token=" ".join(word for word in token_)
  return token.strip()
```

```python
clean_token(text),text
```

```python
data['clean_text']=data.clean_text.apply(clean_token)
```
